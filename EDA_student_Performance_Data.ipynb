```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis as LDA


# Assuming your data is in a DataFrame called df
X = df_featured.drop('Class', axis=1)
y = df_featured['Class']


# Standardize the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Apply LDA
lda = LDA(n_components=2)  # Reducing to 2 components for 2D visualization
X_lda = lda.fit_transform(X_scaled, y)

# Convert the LDA result to a DataFrame for easier plotting
lda_df = pd.DataFrame(X_lda, columns=['LDA1', 'LDA2'])
lda_df['Class'] = y

# Plotting the LDA result
plt.figure(figsize=(10, 7))
classes = np.unique(y)

# Plot each class separately
for cls in classes:
    plt.scatter(lda_df.loc[lda_df['Class'] == cls, 'LDA1'],
                lda_df.loc[lda_df['Class'] == cls, 'LDA2'],
                label=f'Class {cls}')

plt.title('LDA Visualization')
plt.xlabel('LDA1')
plt.ylabel('LDA2')
plt.legend()
plt.grid(True)
plt.show()
```
